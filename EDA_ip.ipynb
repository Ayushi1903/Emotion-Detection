```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
sns.set(color_codes=True)
```

```python
train_directory = "train/"
train_dir = os.listdir(train_directory)
classes = []

for folder in train_dir:
    classes.append(folder)
    print(folder)
```

```python
train_counts = []

for folder in train_dir:
    class_path = train_directory + folder + "/"
    list_train = []
    count = 0
    for file in os.listdir(class_path):
        count +=1
    
    train_counts.append(count)
    
train_counts
```

BAR GRAPH

```python
plt.bar(classes, train_counts, width=0.5)
plt.title("Bar Graph of Train Data")
plt.xlabel("Classes")
plt.ylabel("Counts")
```

SCATTER PLOT

```python
plt.scatter(classes, train_counts)
plt.plot(classes, train_counts, '-o')
plt.show()
```

DISPLOT

```python
x = np.array(train_counts)
sns.distplot(x, kde = True, bins = 5)

sns.distplot(x)
```

Reading and analysing the images

```python
for folder in train_dir:
    class_path = train_directory + folder + "/"
    for file in os.listdir(class_path):
        if file.endswith(".jpg"):
            final_path = class_path + file
            print(final_path)
            img = cv2.imread(final_path, cv2.IMREAD_UNCHANGED)
        break
    break
```

```python
plt.imshow(img)
```

```python
height, width = img.shape

print("""
The Height of the image = {0} 
The Width of the image = {1} 
The Number of channels = {2}""".format(height, width, 1))
```

```python
image_paths = []

for folder in train_dir:
    class_path = train_directory + folder + "/"
    for file in os.listdir(class_path):
        if file.endswith(".jpg"):
            final_path = class_path + file
            image_paths.append(final_path)
        break
```

```python
image_paths
```

```python
import matplotlib.image as mpimg

for file in image_paths:
    image = mpimg.imread(file)
    plt.figure()
    pos1 = file.find('/')
    pos2 = file.find('/0')
    title = file[pos1+1:pos2]
    plt.title(title, fontsize = 20)
    plt.imshow(image)
```
